```python
train = np.load('/home/moucheng/projects_codes/mnist_seg/data/np/train/raw.npy')
```

```python
def display(rows, columns, images, values=[], predictions=[]):
    fig = plt.figure(figsize=(9, 11))

    ax = []

    for i in range(columns*rows):
        img = images[i]
        ax.append(fig.add_subplot(rows, columns, i+1))
        plt.imshow(img, cmap='gray', vmin=0, vmax=1)

    plt.show()

r = 3
c = 3
# idx = np.random.randint(1, 1000, size=r*c)
idx = np.arange(r*c)

images = train[idx,:]
images = images[:,0]

display(rows=r, columns=c, images=images, predictions=[])
```
